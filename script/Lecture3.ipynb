```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 p 的取值范围
p_values = np.linspace(0, 1, 100)

# 计算第一个方程的结果
result1 = 5 * p_values + 0 * (1 - p_values)

# 计算第二个方程的结果
result2 = 4 * p_values + 2 * (1 - p_values)

# 计算第三个方程的结果
result3 = p_values + 4 * (1 - p_values)

# 绘制图像
plt.plot(p_values, result1, label='5p + 0(1-p)')
plt.plot(p_values, result2, label='4p + 2(1-p)')
plt.plot(p_values, result3, label='p + 4(1-p)')
plt.xlabel('p')
plt.ylabel('Expected Payoff')
plt.title('Graphs of 5p + 0(1-p), 4p + 2(1-p), and p + 4(1-p)')
plt.legend()
plt.grid(True)
plt.show()
```
